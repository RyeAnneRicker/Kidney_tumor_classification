```python
# Because some features were found to be of non-Normal distribution by the Shapiro-Wilkes test, the testing of
# individual features was done by the Mann Whitney Test, which does not have to assume Normality
```

```python
# import Libraries
from scipy.stats import mannwhitneyu

import pandas as pd
import numpy as np
```

```python
# import the data
df = pd.read_excel('/Users/RyeAnne/Documents/Reseach/KidneyCancer/FeatureMatrices/FM_Grade_NC.xls')
m,n = df.shape  # used to make a storage list of length n
print(n)
df.head()
```

```python
# make list to store the values in
mw_p = df.T  # transpose original matrix to get column header to be row headers
mw_p = mw_p.iloc[:,0] # get rid of all but the first column, which will be used for storage
mw_p = mw_p.drop(['label'], axis = 0) # drop label column
```

```python
# split data into groups 
g2 = df[df['label']==2] # set first group
g2 = g2.drop(['label'], axis = 1) # drop first column
g3 = df[df['label']==3] # set second group
g3 = g3.drop(['label'], axis = 1) # drop first row

g3.head()
```

```python
for i in range(g2.shape[1]):  # loop through all columns of the data
    stat, p = mannwhitneyu(g2.iloc[:,i],g3.iloc[:,i])  # calculate the mwu stat and p value
    mw_p[i] = p  # place the p-value into the mw vector
```

```python
print(mw_p)
mw_p.to_excel('/Users/RyeAnne/Documents/Reseach/KidneyCancer/FeatureMatrices/NC_Grade_MannWhitneyUTest.xls') 
```
